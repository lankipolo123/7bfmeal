```python
# ============================================================================
# TRAIN NEW CLASSES (Fallen Barrier + Fallen Cones) ON EXISTING MODEL - FAST
# ============================================================================

from google.colab import drive
drive.mount('/content/drive')

!pip install ultralytics roboflow

from ultralytics import YOLO
from roboflow import Roboflow
from google.colab import files
import shutil
import torch

print(f"GPU: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU'}")

# ============================================================================
# UPLOAD YOUR EXISTING ROADBLOCK MODEL
# ============================================================================

print("\nğŸ“¤ Upload your existing roadblocks .pt model...")
uploaded = files.upload()
existing_model = list(uploaded.keys())[0]
print(f"âœ… Uploaded: {existing_model}")

# ============================================================================
# DOWNLOAD NEW DATASET (3K images with Fallen Barrier + Fallen Cones)
# ============================================================================

print("\nğŸ“¥ Downloading NEW dataset (Fallen Barriers/Cones)...")

rf = Roboflow(api_key="SzttdelfmuWaCwAz2N5u")
project = rf.workspace("dequillaprojects").project("roadblocks2-npfaj")
dataset = project.version(1).download("yolov11")

dataset_path = dataset.location
print(f"âœ… Dataset: {dataset_path}")

# ============================================================================
# TRAIN ON NEW CLASSES (TRANSFER LEARNING FROM YOUR MODEL)
# ============================================================================

print("\nğŸš€ Training NEW classes on your existing model...")
print("   This adds Fallen_Barrier and Fallen_Cones to your model!")

model = YOLO(existing_model)  # Load YOUR model

results = model.train(
    data=f'{dataset_path}/data.yaml',
    epochs=50,              # Fast training - only 50 epochs
    imgsz=640,
    batch=32,
    lr0=0.0001,             # Lower LR for fine-tuning
    optimizer='AdamW',
    patience=15,

    name='roadblocks_with_new_classes',
    project='/content/drive/MyDrive',
    device=0,
    cache=True
)

print("\nâœ… Training complete!")

# ============================================================================
# EXPORT TO TFLITE FP32
# ============================================================================

print("\nğŸ“¦ Exporting to TFLite FP32...")

best_model = '/content/drive/MyDrive/roadblocks_with_new_classes/weights/best.pt'
final_model = YOLO(best_model)

# Export
tflite_path = final_model.export(format='tflite', imgsz=640, int8=False)

# Rename
final_name = 'roadblocks_FP32.tflite'
shutil.copy(tflite_path, f'/content/{final_name}')
shutil.copy(tflite_path, f'/content/drive/MyDrive/{final_name}')

print(f"\nâœ… DONE!")
print(f"ğŸ’¾ Saved: MyDrive/{final_name}")
print(f"ğŸ“± Place in: android/app/src/main/assets/{final_name}")

# Download
print("\nğŸ“¥ Downloading...")
files.download(f'/content/{final_name}')

# Show results
val_results = final_model.val()
accuracy = val_results.box.map50 * 100
print(f"\nğŸ¯ Final Accuracy: {accuracy:.2f}%")
```
